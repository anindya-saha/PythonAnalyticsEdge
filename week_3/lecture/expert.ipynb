Back to main problem..Creating a confusion matrix

```python
pd.crosstab(quality_train.PoorCare, predict_train>0.5)
```

```python
# pred_table[i,j] refers to the number of times “i” was observed and the model predicted “j”.
# Correct predictions are along the diagonal.

def conf_info(prediction_df, t=0.5):
    conf = prediction_df.pred_table(t)
    print(conf)
    print('Specificity  {0}'.format(conf[0,0]/sum(conf[0])))
    print('Sensitivity {0}'.format(conf[1,1]/sum(conf[1])))
```

```python
conf_info(quality_log,0.5)
```

```python
conf_info(quality_log,0.7)
```

```python
conf_info(quality_log,0.2)
```
