```python
independent_variables = framingham.columns.drop("TenYearCHD").tolist()
```

```python
independent_variables
```

```python
formula_string = "TenYearCHD ~ "+" + ".join(independent_variables)
```

```python
formula_string
```

```python
# Don't know how to do it with only "."
framingham_log = sm.Logit.from_formula(formula_string, train).fit()
```

```python
framingham_log.summary()
```

```python
predict_test = framingham_log.predict(test.dropna())
```

```python
len(predict_test)
```

```python
len(test.dropna())
```

```python
predict_test
```

```python
def confusion(df,attribute,prediction,threshold=0.5):
    
    return pd.crosstab(df[attribute], prediction>threshold)
```

```python
confusion(test.dropna(),"TenYearCHD",predict_test)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
fpr, tpr, thresholds = roc_curve(test.dropna().TenYearCHD, predict_test)
```

```python
auc(fpr,tpr)
```
