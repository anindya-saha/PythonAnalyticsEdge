```python
nba = pd.read_csv("data/NBA_train.csv")
```

```python
nba.dtypes
```

Renaming variable starting with numbers as statsmodels does not like them

```python
nba = nba.rename(columns={'2P':'X2P','2PA':'X2PA','3P':'X3P','3PA':'X3PA'})
```

```python
nba.describe()
```

```python
nba
```

Number of games need to win to reach playoffs

Viewing wins and playoffs

```python
pd.crosstab(rows=nba.W, cols=nba.Playoffs)
```

Teams winning 35 or fewer games almost never make it to the finals

```python
nba["pts_diff"] = nba.PTS - nba.oppPTS
```

Scatter plots between points difference and wins

```python
nba.plot(kind="scatter",x="pts_diff",y="W")
```

Creating a linear model for wins registered vs pts difference

```python
wins_reg = sm.ols("W~pts_diff",nba).fit()
```

```python
wins_reg.summary()
```

Regression equation is : $W = 41 + 0.0326*pts\_diff$

Creating a model for number of points

```python
points_reg = sm.ols("PTS~ X2PA+ X3PA+FTA+AST+ORB+DRB+TOV+STL+BLK", nba).fit()
```

```python
points_reg.summary()
```

Finding the residuals

```python
points_reg.resid
```

Finding SSE

```python
sse = (points_reg.resid**2).sum()
```

```python
sse
```

```python
import numpy as np
```

```python
rmse = np.sqrt(sse/len(nba.index))
```

```python
rmse
```

Removing turnovers from the model

```python
points_reg2 = sm.ols("PTS~ X2PA+ X3PA+FTA+AST+ORB+DRB+STL+BLK", nba).fit()
```

```python
points_reg2.summary()
```

Removing differenece of rebounds

```python
points_reg3 = sm.ols("PTS~ X2PA+ X3PA+FTA+AST+ORB+STL+BLK", nba).fit()
```

```python
points_reg3.summary()
```

Removing blocks

```python
points_reg4 = sm.ols("PTS~ X2PA+ X3PA+FTA+AST+ORB+STL", nba).fit()
```

```python
points_reg4.summary()
```

Now we find SSE and RMSE again

```python
sse_4 = (points_reg4.resid**2).sum()
```

```python
rmse_4 =np.sqrt(sse_4/len(nba.index))
```

```python
rmse_4
```

Making predictions on test data

Finding out of sample error

```python
sse= ((nba_test.PTS-points_reg4.predict(nba_test))**2).sum()
```

```python
sst =((nba_test.PTS-nba.PTS.mean())**2).sum()
```

```python
r2 = 1-sse/sst
```

```python
r2
```
